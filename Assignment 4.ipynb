```python
print("1. Addition","\n" "2. Subtraction","\n" "3. Multiplication","\n" "4. Division","\n" "5. Power")
select = int(input("Select what you want to do: "))
## Addidtion
if(select == 1):
    def add(a,b):
        return(a+b)
    a = int(input("Enter First number: "))
    b = int(input("Enter Second number: "))
    c = add(a,b)
    print("Answer: ",c)
    
## Subtraction
if(select == 2):
    def subt(a,b):
        return(a-b)
    a = int(input("Enter First number: "))
    b = int(input("Enter Second number: "))
    c = subt(a,b)
    print("Answer: ",c)
##Multiplication
if(select == 3):
    def multi(a,b):
        return(a*b)
    a = int(input("Enter First number: "))
    b = int(input("Enter Second number: "))
    c = multi(a,b)
    print("Answer: ",c)
##Division
if(select == 4):
    def divi(a,b):
        while(a == 0):
            print("Denominator should not be 0")
            a = int(input("Enter Denominator again: "))
        return(b/a)
    b = int(input("Enter Numinator: "))
    a = int(input("Enter Denominator: "))
    c = divi(a,b)
    print("Answer: ",c)

##Power
if(select == 5):
    def powe(a,b):
        return(a**b)
    b = int(input("Enter Power: "))
    a = int(input("Enter Number: "))
    c = powe(a,b)
    print("Answer: ",c)
```

```python
list = ["ABC",2,75,3,2,"CBA",45,4,64,2]
X = 0
for i in range(len(list)):
    if(type(list[i]) == int):
        X += list[i]
print("Sum of all numeric values in list is: ",X)
```

```python
list = ["ABC",2,75,3,2,"CBA",45,3,4,75,"ABC",4,64,2]
for i in range(len(list)):
    count = 0
    for j in range(len(list)):
        if(list[i] == list[j]):
            count += 1
    if(count>1):
        print(list[i]," is Repeated",count,"times")
```
