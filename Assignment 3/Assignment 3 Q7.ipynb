```python
English=float(input("Enter your marks: "))
Math=float(input("Enter your marks: "))
Stats=float(input("Enter your marks: "))
Islamiat=float(input("Enter your marks: "))
Computer=float(input("Enter your marks: "))
Total=500
percentage=((English+Math+Stats+Islamiat+Computer)/Total)*100
if percentage > 100 and percentage <0:
    print("you make a mistake in inputs")
elif percentage <100 and percentage >= 80:
    print("Grade A+")
elif percentage <80 and percentage >= 70:
    print("Grade A")
elif percentage <70 and percentage >= 60:
    print("Grade B")
elif percentage <60 and percentage >= 50:
    print("Grade C")
elif percentage <50 and percentage >= 40:
    print("Grade D")
elif percentage <40 and percentage >= 33:
    print("Grade E")
else:
    print("FAIL")
```
